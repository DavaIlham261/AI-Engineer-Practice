Berapa rata-rata skor total (overall score) universitas di setiap negara

```python
score = times_data.copy()
score["total_score"] = score.total_score.replace("-", np.nan).astype("float")
avg_score = score.groupby(["year", "country"]).total_score.mean().fillna(0)

def fill_score(row):
    if pd.isna(row["total_score"]):
        try:
            return avg_score.loc[(row["year"], row["country"])]
        except KeyError:
            return np.nan
    return row["total_score"]

score["total_score"] = score.apply(fill_score, axis=1)
score.groupby(['country', 'year']).total_score.mean().unstack(fill_value=0).sort_values(by=2016, ascending=False)

# avg_score.unstack(fill_value=0)
# score["total_score"] = score.total_score.fillna()

# score["total_score"] = score.total_score.astype("float")

# score.groupby("year").total_score.mean()
# times_data["total_score"] = times_data["total_score"].fillna()
# times_data["total_score"] = times_data["total_score"].astype("float")
# times_data["total_score"].value_counts(dropna=False).sort_index(ascending=False)
```
