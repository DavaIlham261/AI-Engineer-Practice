### SUMMARY FUNCTION AND MAPS

```python
wine_review
```

```python
wine_review.points.describe()
wine_review.taster_name.describe()
```

```python
wine_review.points.mean()
wine_review.taster_name.unique()
wine_review.taster_name.value_counts()
```

```python
review_point_mean = wine_review.points.mean()
wine_review.points.map(lambda p: p - review_point_mean)
```

```python
def remean_points(row):
    row.points = row.points - review_point_mean
    return row

wine_review.apply(remean_points, axis='columns')
```

```python
wine_review.head(1)
```

```python
review_point_mean = wine_review.points.mean()
wine_review.points - review_point_mean
```

```python
wine_review.country + " - " + wine_review.region_1
```

```python
max_bargain = (wine_review.points / wine_review.price).idxmax()
wine_review.loc[max_bargain, "title"]
```

```python
def assign_stars(row):
    if row['points'] >= 95 or row['country'] == "Canada":
        return "***"
    elif row['points'] >= 85:
        return "**"
    else:
        return "*"

wine_review.apply(assign_stars, axis=1)
```

### GROUPING AND SORTING

```python
wine_review.groupby('points').points.count()
wine_review.groupby('points')['price'].min()
wine_review.groupby('winery').apply(lambda df: df.title.iloc[0])
wine_review.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
wine_review.groupby(['country', 'province']).apply(lambda df: df.loc[df.price.fillna(-1).idxmax()]) # tidak boleh pake idxmax karena ada NaN
wine_review.groupby(['country']).price.agg([len, min, max])
wine_review.groupby('country').agg({'price': 'min', 'points': 'max'})
wine_review.isna().sum()
```

```python
countries_reviewed = wine_review.groupby(['country', 'province']).description.agg([len])
countries_reviewed
mi = countries_reviewed.index
type(mi)
countries_reviewed.reset_index()
```

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
countries_reviewed.sort_values(by='len', ascending=False)
countries_reviewed.sort_index()
countries_reviewed.sort_values(by=['country', 'len'])
```

```python
wine_review.groupby('taster_twitter_handle').size().sort_values(ascending=False)
wine_review.groupby('taster_twitter_handle').taster_twitter_handle.count().sort_index(ascending=True)
```

```python
price_extremes = wine_review.groupby('variety')['price'].agg([min, max])
sorted_varieties = price_extremes.sort_values(['min', 'max'], ascending=False)

reviewer_mean_ratings = wine_review.groupby('taster_name').points.mean()
print(reviewer_mean_ratings)
reviewer_mean_ratings.describe()

wine_review

```
